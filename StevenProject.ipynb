```python
import pandas as pd
from simple_salesforce import Salesforce, SalesforceAuthenticationFailed

# Salesforce authentication credentials
api_key = 'Z4dBaPUfH6FJFt6OaAZ8jHcq'

# Connect to Salesforce
try:
    sf = Salesforce(instance_url='http://trade.my.salesforce.com/services/data/v57.0/', session_id=api_key)
except SalesforceAuthenticationFailed as e:
    print(f"Failed to authenticate with Salesforce. {e}")
    exit()

# Read the names from the Excel file
df = pd.read_excel('US Companies.xlsx')
names = df['Company Name'].tolist()

# Search for the profiles on Salesforce
for name in names:
    try:
        results = sf.query(f"SELECT Id FROM User WHERE LastName = '{name}'")
        if results['records']:
            print(f"{name} has a profile on Salesforce.")
        else:
            print(f"{name} does not have a profile on Salesforce.")
    except Exception as e:
        print(f"An error occurred while searching for {name} on Salesforce. {e}")
```
