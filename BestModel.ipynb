```python
import tkinter as tk
from tkinter import filedialog
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import load_model
import numpy as np
from PIL import ImageTk, Image

# Load the default model
model_file_name = "/content/best_model_100threshold_6_good.h5"
model = load_model(model_file_name)

# Create a Tkinter window
window = tk.Tk()
window.title("Image Classifier")

# Set the window size and background color
window.geometry("500x400")
window.config(bg="#F9F9F9")

# Create a label for the input image
image_label = tk.Label(window, text="Select an image to classify", font=("Helvetica", 16), fg="#333333", bg="#F9F9F9")
image_label.pack(pady=20)

# Create a button to select the input image
def load_image_file():
    global img, image_name
    image_file = filedialog.askopenfilename(title="Select Image", filetypes=(("Image files", "*.jpg;*.jpeg;*.png;*.bmp;*.gif"), ("All files", "*.*")))
    if image_file:
        image_name = image_file.split("/")[-1]
        img = Image.open(image_file)
        img = img.resize((128, 128))
        img = np.array(img)
        img = np.expand_dims(img, axis=0)
        img = img / 255.0
        predictions = model.predict(img)
        predicted_label.configure(text="Predicted label: " + str(np.argmax(predictions, axis=1)[0]), font=("Helvetica", 14), fg="#333333", bg="#F9F9F9")
        image_name_label.configure(text="Selected image: " + image_name, font=("Helvetica", 12), fg="#333333", bg="#F9F9F9")

image_button = tk.Button(window, text="Select Image", font=("Helvetica", 14), fg="#FFFFFF", bg="#333333", activebackground="#555555", command=lambda: load_image_file())
image_button.pack(pady=10)

# Create a label to display the selected image name
image_name_label = tk.Label(window, text="", font=("Helvetica", 12), fg="#333333", bg="#F9F9F9")
image_name_label.pack(pady=10)

# Create a button to reset the GUI
def reset_gui():
    global img, image_name
    predicted_label.configure(text="", fg="#333333", bg="#F9F9F9")
    image_name_label.configure(text="", fg="#333333", bg="#F9F9F9")
    img = None
    image_name = None

reset_button = tk.Button(window, text="Reset", font=("Helvetica", 14), fg="#FFFFFF", bg="#333333", activebackground="#555555", command=lambda: reset_gui())
reset_button.pack(pady=10)

# Create a label to display the predicted label
predicted_label = tk.Label(window, text="", font=("Helvetica", 14), fg="#333333", bg="#F9F9F9")
predicted_label.pack(pady=10)

# Run the Tkinter event loop
window.mainloop()
```
